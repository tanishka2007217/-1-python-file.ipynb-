```python
# positional argument
```

```python
def show(name,age):
  print(name,age)

show("Tanishka","20")
```

```python
# Keyword argument
```

```python
def show(name,age,marks):
  print(name,age,marks)

  show("Tanishka",18,19)
```

```python
# ** kwargs (key-value)
```

```python
def info(**data):
  print(data)

info(name="Tanishka", age=20)
```

```python
# **args (multiple values)
```

```python
def add(*nums):
  print(sum(nums))

add(10,20,30)
```

```python
def multiplication (*arg):
  multi = 1
  for number in arg:
     multi *= number
     return multi

     mul(10,20,30)
```
