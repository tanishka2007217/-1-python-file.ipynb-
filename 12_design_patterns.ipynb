```python
# Singleton Pattern
# Ensures only one instance of a class exists.
```

```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
```

```python
obj1 = Singleton()
obj2 = Singleton()
```

```python
obj1 is obj2
```

```python
# Factory Pattern
# Creates objects without specifying exact class
```

```python
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"

class AnimalFactory:
    def get_animal(self, animal_type):
        if animal_type == "dog":
            return Dog()
        elif animal_type == "cat":
            return Cat()
```

```python
factory = AnimalFactory()
```

```python
animal = factory.get_animal("dog")
```

```python
animal.speak()
```

```python
# Abstract Factory Pattern
# Creates families of related objects
```

```python
class Car:
    def drive(self):
        pass

class BMW(Car):
    def drive(self):
        return "Driving BMW"

class Audi(Car):
    def drive(self):
        return "Driving Audi"

class CarFactory:
    def get_car(self, brand):
        if brand == "bmw":
            return BMW()
        elif brand == "audi":
            return Audi()
```

```python
factory = CarFactory()
```

```python
car = factory.get_car("bmw")
```

```python
car.drive()
```

```python
# Builder Pattern
# Builds complex objects step by step
```

```python
class House:
    def __init__(self):
        self.parts = []

    def add(self, part):
        self.parts.append(part)

class Builder:
    def build(self):
        house = House()
        house.add("Walls")
        house.add("Roof")
        return house
```

```python
builder = Builder()
```

```python
house = builder.build()
```

```python
house.parts
```

```python
# Prototype Pattern
# Creates new objects by copying existing ones
```

```python
import copy

class Prototype:
    def __init__(self, value):
        self.value = value
```

```python
obj1 = Prototype([1, 2, 3])
obj2 = copy.deepcopy(obj1)
```

```python
obj2.value.append(4)
```

```python
obj1.value
```

```python
obj2.value
```

```python
# Adapter Pattern
# Makes incompatible interfaces work together
```

```python
class OldSystem:
    def old_method(self):
        return "Old method"

class Adapter:
    def __init__(self, obj):
        self.obj = obj

    def new_method(self):
        return self.obj.old_method()
```

```python
old = OldSystem()
```

```python
adapter = Adapter(old)
```

```python
adapter.new_method()
```

```python
# Decorator Pattern
# Adds functionality dynamically
```

```python
def decorator(func):
    def wrapper():
        return "Hello " + func()
    return wrapper

@decorator
def say():
    return "World"
```

```python
say()
```

```python
# Facade Pattern
# Provides simple interface to complex system
```

```python
class CPU:
    def start(self):
        print("CPU started")

class Memory:
    def load(self):
        print("Memory loaded")

class Computer:
    def start(self):
        cpu = CPU()
        memory = Memory()
        cpu.start()
        memory.load()
```

```python
pc = Computer()
```

```python
pc.start()
```

```python
# Proxy Pattern
# Controls access to an object
```

```python
class RealObject:
    def request(self):
        return "Real Request"

class Proxy:
    def __init__(self):
        self.real = RealObject()

    def request(self):
        return "Proxy: " + self.real.request()
```

```python
proxy = Proxy()
```

```python
proxy.request()
```

```python
# Composite Pattern
# Treats individual and group objects uniformly
```

```python
class Leaf:
    def show(self):
        print("Leaf")

class Composite:
    def __init__(self):
        self.children = []

    def add(self, obj):
        self.children.append(obj)

    def show(self):
        for child in self.children:
            child.show()
```

```python
c = Composite()
```

```python
c.add(Leaf())
```

```python
c.show()
```

```python
# Behavioral Design Patterns
# Deal with object communication
```

```python
# Observer Pattern
# Notifies multiple objects on change
```

```python
class Subject:
    def __init__(self):
        self.observers = []

    def add(self, obs):
        self.observers.append(obs)

    def notify(self):
        for obs in self.observers:
            obs.update()

class Observer:
    def update(self):
        print("Updated!")
```

```python
sub = Subject()
obs = Observer()
```

```python
sub.add(obs)
```

```python
sub.notify()
```

```python
# Strategy Pattern
# Switch algorithms dynamically
```

```python
class StrategyA:
    def execute(self):
        return "Strategy A"

class StrategyB:
    def execute(self):
        return "Strategy B"

class Context:
    def __init__(self, strategy):
        self.strategy = strategy

    def run(self):
        return self.strategy.execute()
```

```python
context = Context(StrategyA())
```

```python
context.run()
```

```python
# Command Pattern
# Encapsulates request as object
```

```python
class Light:
    def on(self):
        print("Light ON")

class Command:
    def __init__(self, light):
        self.light = light

    def execute(self):
        self.light.on()
```

```python
light = Light()
```

```python
cmd = Command(light)
```

```python
cmd.execute()
```

```python
# Iterator Pattern
# Provides sequential access
```

```python
class MyCollection:
    def __init__(self):
        self.items = [1, 2, 3]

    def __iter__(self):
        return iter(self.items)
```

```python
for item in MyCollection():
    print(item)
```

```python
# State Pattern
# Changes behavior based on state
```

```python
class StateA:
    def handle(self):
        return "State A"

class StateB:
    def handle(self):
        return "State B"

class Context:
    def __init__(self, state):
        self.state = state

    def request(self):
        return self.state.handle()
```

```python
ctx = Context(StateA())
```

```python
ctx.request()
```
