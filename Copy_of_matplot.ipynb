```python
# Create sample data

# 'days' represents time (x-axis)
days = [1, 2, 3, 4, 5]

# 'temperature' represents values (y-axis)
temperature = [30, 32, 35, 33, 31]
```

```python
# Create a new figure with a specific size (width=7 inches, height=4 inches)
plt.figure(figsize=(7, 4))

# Plot a line graph
# marker='o' adds circular points on each data value
plt.plot(days, temperature)

# Add a title to the graph
plt.title("Temperature Trend")

# Label the x-axis
plt.xlabel("Days")

# Label the y-axis
plt.ylabel("Temperature (°C)")

# Show grid lines for better readability
plt.grid(True)

# Display the graph
plt.show()
```

```python
# Categorical Plots
```

```python
categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")

plt.show()
```

```python
plt.barh(categories, values)
```

```python
# 8. Pie Chart
```

```python
labels = ["A", "B", "C"]
sizes = [30, 40, 30]
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Simple Pie Chart")
plt.show()
```

```python
# Histogram (Frequency Distribution)
```

```python
scores = [42, 45, 48, 52, 55, 58, 62, 65, 70, 75]

counts, bins, patches = plt.hist(scores, bins=5, edgecolor='black', color='skyblue')
plt.xticks(bins)

plt.bar_label(patches)

plt.title("Histogram of Student Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
# Box Plot
```

```python
scores = [45, 45, 48, 52, 55, 58, 62, 65, 70, 75]

# Use tick_labels for the x-axis label
plt.boxplot(scores, tick_labels=['Student Scores'])

# Set the y-axis limits (bottom, top)
plt.ylim(40, 80)

plt.title("Box Plot of Student Scores")
plt.ylabel("Scores")
plt.show()
```

```python
x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.legend()
plt.show()
```
