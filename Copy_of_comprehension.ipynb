```python
# Comprehension
```

```python
# list comprehension
```

```python
# basic
```

```python
numbers = [1,2,3,4,5]
```

```python
squares =[x*x for x in numbers]
```

```python
print(squares)
```

```python
print(squares)
```

```python
# with condition
```

```python
numbers = [1,2,3,4,5,6,]
```

```python
even_numbers = [x for x in numbers if x % 2 == 0]
```

```python
print(even_numbers)
```

```python
# set comprehension
```

```python
numbers = [1,2,3,4,5,]
```

```python
unique_squares = {x*x for x in numbers}
```

```python
print(unique_squares)
```

```python
# Dictionary comprehension
```

```python
numbers =[1,2,3,4,]
```

```python
square_dict = {x: x*x for x in numbers}
```

```python
print(square_dict)
```

```python
numbers = [1,2,3,4,5]
```

```python
even_dict = {x: x*x for x in numbers if x % 2 == 0}
```

```python
print(even_dict)
```

```python
# Genertor comprehension
```

```python
numbers = [1,2,3,4,5]
```

```python
gen = (x*x for x in numbers)
```

```python
print(gen)
```

```python
for value in gen:
  print(value)
```
