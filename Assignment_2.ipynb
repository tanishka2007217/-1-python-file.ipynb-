```python
#[2]  factorial using function and validation
```

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact*= i
        return fact

n = int(input("Enter number:"))

if n < 0:
   print("Invalid input")
else:
   print("Invalid:",factorial(n))
```

```python
# [3] prime check + factorial count
```

```python
def is_prime(n):
    count = 0
    for i in range(1, n+1):
       if n % i ==0:
           count +=1
           return count ==2,count

n = int(input("Enter number:"))
prime,factors = is_prime(n)
print("prime:",prime)
print("factor count:",factors)
```

```python
# [8] Recursive factorial
```

```python
def fact(n):
    if n==0:
        return 1
        return n* fact(n-1)
print(fact(5))
```

```python
# [9] Recursive digit operations
```

```python
def sum_digits(n):
    if n==0:
        return 1
        return n%10 + sum_digits(n//10)
def reverse(n,rev=0):
    if n ==0:
       return rev
       return reverse(n//10,rev*10 + n%10)
n = int(input())
print(sum_digits(n))
print(reverse(n))
```
